```python
import pandas as pd
url ='https://data.moenv.gov.tw/api/v2/aqx_p_432?api_key=58d6040c-dca7-407f-a244-d0bfdfa8144a&limit=1000&sort=ImportDate%20desc&format=CSV'
df = pd.read_csv(url)
df.head()
```

留下需要的欄位('sitename','county','pollutant','status','aqi')

```python
df = df [['sitename', 'county', 'pollutant', 'status', 'aqi']]
df
```

找出空氣品質不良的測站(aqi>100)

```python
condition = df['aqi']>100
df[condition]
```

查詢空氣汙染指標物(pollutant)有哪些?

```python
df['pollutant'].unique()
```

請找出高雄市的測站資料

```python
condition = df['county'] == '高雄市'
df[condition]
```

計算各縣市的AQI

```python
df.groupby('county')['aqi'].mean().sort_values(ascending=False)
```
